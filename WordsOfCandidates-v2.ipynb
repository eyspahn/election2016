#### "JJ" is an adjective. Let's count the kinds of adjectives the candidates used.

```python
secclintondf.tags[0]
# interesting, "-H" is considered an adjective
```

```python
counter_clinton = Counter()
secclintondf.tags.apply(lambda s: counter_clinton.update(s))
```

```python
for key in sorted(counter_clinton, key=counter_clinton.get, reverse=True):
    print(key,counter_clinton[key])
```

```python
counter_trump = Counter()
mrtrumpdf.tags.apply(lambda s: counter_trump.update(s))
```

```python
for key in sorted(counter_trump, key=counter_trump.get, reverse=True):
    print(key,counter_trump[key])
```
